```python
import pandas as pd
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df_birth_year = pd.read_csv('datasets/birth_year.csv', sep=',')
df_political_leaning = pd.read_csv('datasets/political_leaning.csv', sep=',')
```

```python
print(f"Number of rows in df_birth_year: {len(df_birth_year)}")
print(f"Number of unique authors in df_birth_year: {len(df_birth_year['auhtor_ID'].unique())}")
birth_year_counts = sorted(Counter(df_birth_year['birth_year']).items())
x, y = zip(*birth_year_counts)
plt.figure(figsize=(16,6))
plt.plot(x, y)
plt.title('Birth Year Distribution ')
plt.xlabel('Year of Birth')
plt.ylabel('Number of Rows');
```

```python
print(f"Number of rows in df_political_leaning: {len(df_political_leaning)}")
print(f"Number of unique authors in df_political_leaning: {len(df_political_leaning['auhtor_ID'].unique())}")
print(f"Number of posts for each political leaning: \n{df_political_leaning['political_leaning'].value_counts()}")
```

```python
posts_in_both = len(pd.merge(df_birth_year, df_political_leaning, on=["post", "auhtor_ID"], how="inner"))
print(f"Number of rows with the same 'post' column in both datasets: {posts_in_both}")
```
